```python
#将上面内容整合成函数
def get_charpter_link_list(url):
    headers ={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
             }
    r = requests.get(url,headers = headers)
    r.encoding = r.apparent_encoding
    bs =BeautifulSoup(r.text,'lxml')
    all_a_tag = bs.find('dl').find_all('a')
    charpter_link = []
    for i in all_a_tag:
        charpter_link.append(i['href'])
    return charpter_link
    
```

```python
charpter_link_list = get_charpter_link_list("http://www.shuquge.com/txt/8659/index.html")
```

```python
#整合函数
def get_one_page_content(url):
    headers ={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
             }
    r = requests.get(url,headers = headers)
    r.encoding = r.apparent_encoding
    bs =BeautifulSoup(r.text,'lxml')
    title = bs.find('h1').text
    showtxt = bs.find('div','showtxt').text
    content = showtxt.replace('\xa0','').replace('\r','\n\n')
    return title,content
    
```

```python
get_one_page_content("http://www.shuquge.com/txt/8659/2324752.html")
```

```python
basic_link = "http://www.shuquge.com/txt/8659/"
n=1
for i in charpter_link_list:
    print(basic_link+i)
    title,content = get_one_page_content(basic_link+i)
    f = open(str(n)+'.'+title+'.txt','w')
    f.write(title)
    f.write('\n\n')
    f.write(content)
    f.close()
    n+=1
```

```python
#将功能打包成函数可以调用该网站的任意一本小说
def download_novel(basic_link,novel_name):
    """basic_link:小说主页，到数字为止
       novel_name:小说名称   
    """
    import os
    os.mkdir(novel_name)
    charpter_link_list = get_charpter_link_list(basic_link)
    n=1
    for i in charpter_link_list:
        print(basic_link+i)
        title,content = get_one_page_content(basic_link+i)
        f=open(novel_name+'/'+str(n)+'.'+title+'.txt','w')
        f.write(title)
        f.write('\n\n')
        f.write(content)
        f.close()
        n+=1
```

```python
download_novel('http://www.shuquge.com/txt/5809/','元尊')
```
